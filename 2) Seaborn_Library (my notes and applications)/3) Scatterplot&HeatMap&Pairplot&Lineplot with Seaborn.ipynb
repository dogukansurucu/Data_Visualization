# Korelasyon Grafikleri

> **Korelasyon** : Değişkenler arasındaki ilişkiyi ifade eden istatiksel bir terimdir.

> Bundan önceki grafiklerde tek bir değişken üzerinden veri görselleştirme yapıyorduk bunlara ek olarak eklenen değişkenler boyut anlamında çaprazlama anlamında eklenen değişkenlerdi.

> Şimdi ise iki değişken üzerinden veri görselleştireceğiz. Aynı anda iki sayısal değişkenin ortaya çıkaracağı yapıyı ele alacağız.

## **Scatterplot** (saçılım,serpilme grafiği)

**Scatterplot** ---> **Sayısal değişkenler** arasındaki ilişkiyi gösterir.

total_bill: yemeğin toplam fiyatı (bahşiş ve vergi dahil)

tip: bahşiş

sex: ücreti ödeyen kişinin cinsiyeti (0=male, 1=female)

smoker: grupta sigara içen var mı? (0=No, 1=Yes)

day: gün (3=Thur, 4=Fri, 5=Sat, 6=Sun)

time: ne zaman? (0=Day, 1=Night)

size: grupta kaç kişi var?

```python
?sns.scatterplot
```

```python
import seaborn as sns
tips = sns.load_dataset("tips")
df = tips.copy()
df.head()
```

```python
sns.scatterplot(x = "total_bill", y = "tip", data = df);

# toplam ödenen ücret arttıkça bahşişlerde artma gözlemlenmiş
```

## Korelasyon Çaprazlamalar

```python
sns.scatterplot(x = "total_bill", y = "tip", hue = "time", data = df);
```

```python
sns.scatterplot(x = "total_bill", y = "tip", hue = "time", style = "time", data = df);
```

```python
sns.scatterplot(x = "total_bill", y = "tip", hue = "day", style = "time", data = df);
```

```python
sns.scatterplot(x = "total_bill", y = "tip", size = "size", style = "time", data = df);

# size değişkeni sembolünün büyük olması masadaki kişi sayısının fazla olduğu anlamı gelir.
```

```python
sns.scatterplot(x = "total_bill", y = "tip", hue = "size", size = "size", data = df);
```

## **Doğrusal İlişkinin Gösterilmesi**

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")
df = tips.copy()
df.head()

sns.lmplot(x = "total_bill", y = "tip", data = df);                                   # lm(linear Model)plot
```

```python
sns.lmplot(x = "total_bill", y = "tip", hue = "smoker", col = "time", data = df);

# col argümanına göre grafik bölünür. (col argümanındaki değer kaça ayrılmışsa o kadar bölünür.)
```

```python
sns.lmplot(x = "total_bill", y = "tip", hue = "smoker", col = "time", row = "sex", data = df);

# col sütun bazında
# row satır bazında
# iki sayısal değişkenin erkek cinsiyetinde lunch ve dinner değişkeninde grafik nasıl
# iki sayısal değişkenin kadın cinsiyetinde lunch ve dinner değişkeninde grafik nasıl onu görmek istedik.
```

## **Isı Haritası (HeatMap)**

> **Isı haritası, renk kodlarını kullanarak verilerin dikdörtgen şeklinde çizimidir. Parametre olarak bir 2D veri kümesi alır. Bu veri kümesi bir ndarray'e zorlanabilir.**

**VERİ SETİ HİKAYESİ**

```python
import seaborn as sns
flights = sns.load_dataset("flights")
df = flights.copy()
df.head()
# yıllara ve aylara göre yolculuk sayılarını veren 3 değişkenli veri seti
```

```python
df.shape
```

```python
df["passengers"].describe()
```

```python
#sns.heatmap(df)    ---> hata verir.

# çünkü bu yapıdaki veri seti heatmap'e uygun değil. daha iyi bir yapıya dönüştürmemiz gerekiyor.
# bunun için pivot kullanıyoruz.
```

```python
print(df,"\n\n")
```

```python
df = df.pivot("month", "year", "passengers")
sns.heatmap(df);
```

```python
sns.heatmap(df, annot = True, fmt = "d", linewidths = .5);

# annot = true ve fmt = "d" diyerek grafiği daha iyi gözlemlemek için hücrelerin içini passengers verileriyle doldurmuş olduk
# linewidths ile hücreleri bölmüş olduk.
```

```python
# bu grafik yıllar geçtikçe yolcu sayısının arttığını gösteriyor.
# aylar bazında baktığımızda da açık renkteki bölgeler haziran ve temmuz aylarındaki verilerdir.
```

> **pivot_table**, pivotlanmış bir dizin / sütun çifti için yinelenen değerleri işleyebilen pivotun genelleştirilmesidir. Özellikle, anahtar sözcük bağımsız değişkeni aggfunc kullanarak pivot_table'a toplama işlevlerinin bir listesini verebilirsiniz.
pivot_table, pivot tablonun dizini ve sütunu için birden çok sütun kullanmayı da destekler. Hiyerarşik bir dizin sizin için otomatik olarak oluşturulur.

> **pivot_table**, sayısal türlere yalnızca "değerler=" olarak izin verirken, **pivot** dize türlerini "değerler=" olarak alır.

> **pivot()** indeks için bir liste kabul etmez.
**pivot_table()** kabul eder.

## **Çizgi Grafiği**

**VERİ SETİ HİKAYESİ**

```python
import seaborn as sns
fmri = sns.load_dataset("fmri")
df = fmri.copy()
df.head()
```

```python
df["timepoint"].describe()
```

```python
df.groupby("timepoint")[["signal"]].count()

# demekki her bir zaman noktasına göre eşit sayıda sinyal toplanmış
```

```python
df.groupby("timepoint")[["signal"]].describe()
```

**Çizgi Grafik Ve Çaprazlamalar**

```python
import seaborn as sns
fmri = sns.load_dataset("fmri")
df = fmri.copy()
df.head()
sns.lineplot(x = "timepoint", y = "signal", data = df);
```

```python
import seaborn as sns

fmri = sns.load_dataset("fmri")
df = fmri.copy()
df.head()

sns.lineplot(x = "timepoint",
             y = "signal",
             hue = "region",
             style = "event",
             markers = True,
             dashes = False,
             data = df);


#regionı parietal olanların zaman noktası 10'a kadar(yaklaşık) daha fazla sinyal verdiği, 
#zaman noktası 10'dan sonra ise regionı frontal olanların daha fazla sinyal verdiği görülüyor.
```
