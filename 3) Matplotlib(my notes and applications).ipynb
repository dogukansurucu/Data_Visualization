## Matplotlib Çizimi

```python
xeksen = np.array([1,2,3,4])
yeksen = np.array([3,6,9,12])
plt.plot(xeksen,yeksen)
plt.xlabel("x ekseni")
plt.ylabel("y ekseni")
plt.title("Başlık")
plt.grid()
plt.show()
```

```python
xeksen = np.array([1, 2, 6, 8])
yeksen = np.array([3, 8, 1, 10])
plt.xlabel("x ekseni")
plt.ylabel("y ekseni")
plt.title("azalan artan bir grafik")
plt.plot(xeksen,yeksen)
plt.show()
```

```python
x1 = np.array([0, 4, 6.5, 11])
y1 = np.array([2, 6, 4, 8])
x2 = np.array([0, 4, 6, 9])
y2 = np.array([6, 2, 7, 9])

plt.plot(x1, y1, x2, y2)
plt.grid()
plt.show()
```

## **Çizgisiz Çizim**

```python
xeksen = np.array([1, 8])
yeksen = np.array([3, 10])
plt.plot(xeksen, yeksen, 'o')
plt.show()
```

## **Varsayılan x- Değerleri**

```python
"""
Eğer x eksenindeki noktaları belirtmezsek, y noktalarının uzunluğuna bağlı olarak
0, 1, 2, 3, (vb.) varsayılan değerlerini alacaklardır.
"""

a = np.array([8, 1, 10, 5,9])

plt.xlabel("x ekseni")
plt.ylabel("y ekseni")
plt.title("azalan artan bir grafik")
plt.plot(a)
plt.show()
```

## **Markers (İşaretçiler)**

```python
"""
marker: Her noktayı belirli bir işaretleyiciyle vurgulamak için anahtar kelime bağımsız değişkenini kullanabilirsiniz.

"""
yeksen = np.array([3, 8, 1, 10])

plt.plot(yeksen, marker = 'o')
plt.show()
```

**Marker(İşaret) Referansı**

Bu işaretçilerden herhangi birini seçebilirsiniz:

* 'o' **:** Circle
* '*' **:** Star
* '.' **:** Point
* ','**:** Pixel
* 'x'**:** X
* 'X' **:** X (filled)
* '+' **:** Plus
* 'P' **:** Plus (filled)
* 's' **:** Square
* 'D' **:** Diamond
* 'd' **:** Diamond(thin)
* 'p' **:** Pentagon 
* 'H' **:** Hexagon
* 'h' **:** Hexagon
* 'v' **:** Triangle Down
* '^' **:** Triangle Up
* '<' **:** Triangle Left
* '>' **:** Triangle Right
* '1' **:** Tri Down
* '2' **:** Tri Up
* '3' **:** Tri Left
* '4' **:** Tri Right
* '|'**:** Vline
* '_' **:** Hline


```python
yeksen = np.array([3, 8, 1, 10])

plt.plot(yeksen, marker = '*')
plt.show()
```

## **Dizeleri Biçimlendirme**

fmt şu sözdizimi ile yazılır: **marker|line|color**

**Line(hat) Referansı**
* '-' **:** Düz Çizgi
* ':' **:** Noktalı Çizgi
* '--' **:** Kesik Çizgi
* '-.' **:** Kesikli/Noktalı Çizgi

**Color(renk) Referansı**
* 'r' **:** Kırmızı
* 'g' **:** Yeşil
* 'b' **:** Mavi
* 'c' **:** Cyan
* 'm' **:** Magenta
* 'y' **:** Sarı
* 'k' **:** Siyah
* 'w' **:** Beyaz

```python
ypoints = np.array([2, 11, 5, 18])

plt.plot(ypoints,'o:g')
plt.show()
```

```python
ypoints = np.array([2, 11, 5, 18])

plt.plot(ypoints,'o')
plt.show()
```

```python
#İşaret Boyutu --> ms argümanı

ypoints = np.array([2, 11, 5, 18])

plt.plot(ypoints,'o:g',ms=15) 
plt.show()
```

```python
#İşaret dış Rengi --> mec argümanı

ypoints = np.array([2, 11, 5, 18])

plt.plot(ypoints,'o:g',ms=15,mec='b')
plt.grid()
plt.show()
```

```python
#İşaret iç rengi --> mfc argümanı

ypoints = np.array([2, 11, 5, 18])

plt.plot(ypoints,'o-g',ms=15,mfc='y') 
plt.show()
```

```python
# işaret hem içi hem dış rengi

ypoints = np.array([2, 11, 5, 18])

plt.plot(ypoints,'o-b',ms=15,mec='k',mfc='k') 
plt.show()
```

```python
# grafiğe grid(ızgara) çizgileri ekleme

ypoints = np.array([2, 11, 5, 18])

plt.plot(ypoints,'o-b',ms=15,mec='k',mfc='k') 
plt.grid()
plt.show()
```

## Matplotlib ile **Pie Grafiği** Oluşturma

```python
?plt.pie
```

[matplotlib.pyplot.pie](https://matplotlib.org/3.5.3/api/_as_gen/matplotlib.pyplot.pie.html)

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([60, 20, 15, 10])
mylabels = ["FENERBAHÇE", "GALATASARAY", "BEŞİKTAŞ", "TRABZONSPOR"]
mycolors = ["blue", "red", "black", "m"]
plt.pie(y, labels = mylabels, colors=mycolors)
plt.legend(title = "FOOTBALL TEAM:")
plt.show() 
```

```python
y = np.array([55, 25, 5, 5])
mylabels = ["Blue", "Orange", "Green", "Red"]
myexplode = [0.1, 0.1, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode, shadow = True)
plt.show() 
```

## Matplotlib ile **Scatter Plot Grafiği** Oluşturma

```python
?plt.scatter
```

[matplotlib.pyplot.scatter](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html)

**!!!  <br> c argümanı sayesinde bir dizi renk kullanarak her nokta için belirli bir renk uygulayabiliriz.**

```python
x = np.array([2,3,8,4,11,17,12,3,4,11,12,9,6])
y = np.array([99,86,87,88,100,99,94,87,94,78,77,85,86])
colors = np.array(["red","green","blue","yellow","green","black","orange","purple","blue","orange","gray","purple","magenta"])

plt.scatter(x, y, c=colors)

plt.show()
```

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

plt.scatter(x, y, c=colors, cmap='viridis')

plt.colorbar()

plt.show()
```
